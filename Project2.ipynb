## Project 2

In this project, I will plot the amount of car crashes in NYC as a time series on one axis against the traffic in NYC. I hypothesize a positive linear correlation

car crashes: https://data.cityofnewyork.us/Public-Safety/Motor-Vehicle-Collisions-Crashes/h9gi-nx95/about_data

traffic: https://data.cityofnewyork.us/Transportation/Automated-Traffic-Volume-Counts/7ym2-wayt/about_data

```python
import plotly.io as pio

pio.renderers.default = "vscode+jupyterlab+notebook_connected"
```

```python
import pandas as pd
import plotly.express as px
```

```python
#Read in the data
crash_data = pd.read_csv('Motor_Vehicle_Collisions_-_Crashes_20241120.csv')
crash_data
```

First, I will change the time to datetime type

```python
crash_data["CRASH DATE"] = pd.to_datetime(crash_data["CRASH DATE"], format = "%m/%d/%Y")
crash_data
```

Then, I will count up the crashes per day

```python
crashes_per_day = crash_data.resample("D", on="CRASH DATE").size().reset_index(name="Crashes")
crashes_per_day
```

Here is a line graph over time of how crashes look

```python
px.line(
    crashes_per_day,
    x="CRASH DATE",
    y="Crashes",
    title="Car Crashes per Day in NYC"
)
```

That is a little messy, so let's change it to per week

```python
crashes_per_week = crash_data.resample("W", on="CRASH DATE").size().reset_index(name="Crashes")
crashes_per_week
```

```python
px.line(
    crashes_per_week,
    x="CRASH DATE",
    y="Crashes",
    title="Car Crashes per Week in NYC"
)
```

Interesting to see the large drop during COVID has not rebounded. Plus, it still looks messy, should we try per month?

```python
crashes_per_month = crash_data.resample("M", on="CRASH DATE").size().reset_index(name="Crashes")
crashes_per_month
```

```python
px.line(
    crashes_per_month,
    x="CRASH DATE",
    y="Crashes",
    title="Car Crashes per Month in NYC"
)
```

That's better. It looks like there is seasonality in car crashes, with more crashes during the summer. Let's now take a look at traffic to see if there is a similar seasonality

```python
traffic_data = pd.read_csv("Automated_Traffic_Volume_Counts_20241120.csv")
traffic_data
```

This data shows collects "traffic sample volume counts at bridge crossings and roadways" around NYC. We need to add the "Vol" column up by month to see how many cars drove by all of the checkpoints each month. First lets create a new columnn that has the date in one

```python
traffic_data["Date"] = traffic_data["M"] + "/" + traffic_data["D"] + "/" + traffic_data["Yr"]
traffic_data
```

The dates are ints not strings, so we need to make them strings

```python
traffic_data["Yr"] = traffic_data["Yr"].astype(str)
traffic_data["M"] = traffic_data["M"].astype(str)
traffic_data["D"] = traffic_data["D"].astype(str)
```

```python
traffic_data["Date"] = traffic_data["M"] + "/" + traffic_data["D"] + "/" + traffic_data["Yr"]
traffic_data
```

```python
traffic_data["Date"] = pd.to_datetime(traffic_data["Date"], format = "%m/%d/%Y")
traffic_data
```

Now we add up the traffic volume for each month

```python
traffic_vol_per_month = traffic_data.resample("M", on="Date").sum("Vol").reset_index()
traffic_vol_per_month
```

```python
px.line(
    traffic_vol_per_month,
    x="Date",
    y="Vol",
    title="Traffic Volume by Month in NYC"
)
```

Let's make a new data frame that has the above data but starts at Jul 2012, which is when car crash data starts

```python
traffic_vol_per_month2 = traffic_vol_per_month[traffic_vol_per_month["Date"]>="2012-07-31"]
traffic_vol_per_month2
```

```python
px.line(
    traffic_vol_per_month2,
    x="Date",
    y="Vol",
    title="Traffic Volume by Month in NYC"
)
```

We see peaks in April, May, October, and Novemeber, and lulls in December, January, February, July, and August. This could be because school is out for winter break and summer break

Now let's merge the datasets on month

```python
#Rename column
crashes_per_month = crashes_per_month.rename(columns={'CRASH DATE': 'Date'})
#Merge
merged_data = pd.merge(crashes_per_month, traffic_vol_per_month2, on="Date")
merged_data
```

Let's make a scatter plot first

```python
fig = px.scatter(
    merged_data,
    x = "Crashes",
    y = "Vol",
    title = "Car Crashes vs Traffic Volume"
)
fig.show()
```

It looks like there is a weak positive linear association between car crashes and traffic volume. There are two groups of data. The low crash data is post-COVID and the high crash data is pre-COVID. 

Now we can make the line graph with both car crashes and traffic, with car crashes on the left y-axis, traffic on the right y-axis, and months on the x-axis

```python
#The following code was inspired by AI but not directly copied

import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a subplot with secondary y-axis
fig = make_subplots(
    rows=1, cols=1, 
    specs=[[{"secondary_y": True}]],  # Enable secondary y-axis
    shared_xaxes=True  # Share x-axis between the two traces
)

# Plot the first line on the primary y-axis
fig.add_trace(
    go.Scatter(x=merged_data['Date'], y=merged_data['Crashes'], mode='lines', name='Car Crashes'),
    secondary_y=False  # Use primary y-axis
)

# Plot the second line on the secondary y-axis
fig.add_trace(
    go.Scatter(x=merged_data['Date'], y=merged_data['Vol'], mode='lines', name='Traffic Volume'),
    secondary_y=True  # Use secondary y-axis
)

# Update layout with axis labels
fig.update_layout(
    title_text="Car Crashes vs Traffic Volume Over Time",
    xaxis_title="Date",
    yaxis_title="Car Crashes",
    yaxis2_title="Traffic Volume"
)

# Show the figure
fig.show()
```

I would have expected Traffic Volume to fall along with Car Crashes during COVID; however, Traffic Vol seems to have stayed pretty constant although it did fall alittle. That being said, the seasonality looks pretty similar: spikes in Car Crashes corresponded to spikes in Traffic Volume, for the most part. Keep in mind that Traffic Volume measures the amount of cars at certain bridges and checkpoints, not the total amount of cars on the road. Therefore, people may be driving less in general even though the amount of driving in certain areas has only declined by a bit.
