**Clean the data without Pandas**

```python
import csv
data = []
with open("Parking_Data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(row)
data[0]
```

```python
max_rates = [(row['All_Vehi_3']) for row in data]
max_rates
```

*For the values with two numbers, only keep the highest*

```python
max_rates = [x[-5:] for x in max_rates]
max_rates
```

*Do some more cleaning of $ and N/A values*

```python
max_rates = [item.strip() for item in max_rates]
```

```python
max_rates = [item.strip('$') for item in max_rates]
```

```python
max_rates = [item for item in max_rates if item != 'N/A']
```

```python
max_rates = [float(item) for item in max_rates]
```

```python
max_rates = [item for item in max_rates if item != 'nan']
max_rates
```

**Compute mean, median, and mode without Pandas**

```python
average = sum(max_rates) / len(max_rates)
print(f"Mean: {average:.2f}")
```

```python
#Used AI to understand how to define a median function but did not directly copy
def median(data):
    data_sorted = sorted(data)
    n = len(data_sorted)

    if n % 2 == 1: 
        return data_sorted[n // 2]
    else:  
        center1 = data_sorted[n // 2 - 1]
        center2 = data_sorted[n // 2]
        return (center1 + center2) / 2
```

```python
med = median(max_rates)
print(f"Median: {med:.2f}")
```

*The Mode is trickier with pure Python, I will create a dictionary and iterate through it*

*The keys are the data points and the values are the frequency*

```python
frequency = {}
for item in max_rates:
    if item in frequency:
        # If the item is already in the dictionary, increment its frequency
        frequency[item] += 1
    else:
        # If the item is not in the dictionary, add to its frequency
        frequency[item] = 1
frequency
```

*Next, iterate through the dictionary to find the highest frequency data point*

```python
max_frequency = max(frequency.values()) #finds the highest value
mode = 0
for key in frequency:
    if frequency[key] == max_frequency:
        mode = key
print(f"Mode: {mode:.2f}")
#Used AI to understand how to work through mode calcualtion but did not directly copy
```
